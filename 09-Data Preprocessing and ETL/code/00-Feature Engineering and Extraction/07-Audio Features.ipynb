# Audio Features
## Mel Frequency Cepstral Coefficients (MFCCs)
The mel frequency cepstral coefficients (MFCCs) of a signal are a small set of features (usually about 10-20) which concisely describe the overall shape of a spectral envelope. In MIR, it is often used to describe timbre.
<img src='https://miro.medium.com/max/1400/1*dWnjn5LLS0j8St53ACwqSg.jpeg'>

```python
import numpy
import matplotlib.pyplot as plt
import IPython.display
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
import librosa
import librosa.display as ds

x, fs = librosa.load('../dastasets/drum.wav')
ds.waveplot(x, sr=fs)
```

```python
IPython.display.Audio(x, rate=fs)
```

```python
len(x)
```

```python
x
```

```python
fs
```

```python
len(x) / fs
```

```python
mfccs = librosa.feature.mfcc(x, sr=fs)
mfccs.shape
```
