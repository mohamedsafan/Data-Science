**1) Using titanic_full.csv dataset, extract Mr,Miss,Mrs,etc.. from 'Name' column and use it to create new column and call it 'Title', then plot a countplot showing the count for each title in the dataset.**

```python
df = pd.read_csv('../../code/dastasets/titanic_full.csv')
df
```

```python
def extract_title(name):
    if 'Mr.' in name:
        return 'Mr'
    elif 'Miss.' in name:
        return 'Miss'
    elif 'Mrs.' in name:
        return 'Mrs'
    else:
        return 'Untitled'

df['Title'] = df['Name'].apply(extract_title)
df
```

```python
sns.countplot(x='Title', data=df)
```

**2) Using sendy_logistics.csv extract user_id in a seperate column**

```python
df = pd.read_csv('../../code/dastasets/sendy_logistics.csv')
df
```

```python
def extract_user_id(user_id_txt):
    return user_id_txt.split('_')[2]

df['user_id'] = df['User Id'].apply(extract_user_id)
df
```
