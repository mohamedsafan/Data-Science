```python
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
df = pd.read_csv('Pokemon.csv', encoding='latin1')
df
```

```python
df.info()
```

```python
df = df.convert_dtypes()
df.info()
```

```python
sns.distplot(df['Total'], kde=False)
```

```python
sns.kdeplot(df['HP'], shade=True)
```

```python
sns.kdeplot(df['Speed'], shade=True)
```

```python
sns.jointplot(x='Attack', y='Defense', data=df)
```

```python
sns.jointplot(x='HP', y='Speed', data=df, kind='kde')
```

```python
sns.countplot(x='Stage', data=df, palette='viridis')
```

```python
sns.countplot(x='Type 2', data=df, palette='viridis')
```

```python
sns.boxplot(y='Type 1', x='HP', data=df, palette='viridis')
```

```python
sns.boxplot(y='Type 1', x='Attack', data=df, palette='viridis')
```

```python
sns.stripplot(y='Type 1', x='Attack', data=df, palette='viridis')
```

```python
sns.stripplot(y='Speed', x='Stage', data=df, palette='viridis')
```

```python
sns.stripplot(y='Speed', x='Stage', data=df, palette='rocket', hue='Legendary', dodge=True)
```

```python
sns.barplot(x='Type 2', y='HP', data=df, palette='rocket')
```
