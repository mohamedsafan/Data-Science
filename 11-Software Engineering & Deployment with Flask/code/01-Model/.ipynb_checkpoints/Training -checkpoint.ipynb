```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize': [7, 7]}, font_scale=1.2)
```

```python
df = pd.read_csv('bikes.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d', errors="coerce")
df
```

```python
df.info()
```

```python
df['Month'] = df['date'].dt.month
df['Day'] = df['date'].dt.day_name()
df
```

```python
def map_months(x):
    if x in [12, 1, 2]:
        return 'Winter'
    elif x in [3, 4, 5]:
        return 'Spring'
    elif x in [6, 7, 8]:
        return 'Summer'
    elif x in [9, 10, 11]:
        return 'Autumn'

df['Season'] = df['Month'].apply(map_months)
df
```

```python
from datasist.structdata import detect_outliers
```

```python
outliers_indices = detect_outliers(df, 0, ['temperature', 'humidity', 'windspeed'])
len(outliers_indices)
```

```python
outliers_indices
```

```python
df.drop(outliers_indices, inplace=True)
df
```

```python
df = pd.get_dummies(df, columns=['Season'], drop_first=True)
df
```

```python
x = df.drop(['date', 'count', 'Month', 'Day'], axis=1)
x
```

```python
y = df['count']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
models = [LinearRegression(), DecisionTreeRegressor(), RandomForestRegressor(), XGBRegressor()]
```

```python
for model in models:
    model.fit(x_train, y_train)
    print(model.score(x_test, y_test))
```

```python
model = RandomForestRegressor(1000)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
