```python
import requests, json
import time
from datetime import datetime
import csv

with open('weather.csv', 'w') as f:
    writer = csv.DictWriter(f, fieldnames=['temperature', 'wind_speed', 'pressure', 'humidity', 'visibility', 'time_now'])
    writer.writeheader()
    for i in range(10):
        url = "http://api.weatherstack.com/current?access_key=cce43aa3a3de500cc32c5df0e67c9408&query=Cairo"
        response = requests.request("GET", url)
        json_response = json.loads(response.text)
        temperature = json_response['current']['temperature']
        wind_speed = json_response['current']['wind_speed']
        pressure = json_response['current']['pressure']
        humidity = json_response['current']['humidity']
        visibility = json_response['current']['visibility']
        time_now = datetime.now().strftime("%d/%m/%Y, %H:%M:%S")
        writer.writerow({'temperature': temperature, 'wind_speed': wind_speed, 'pressure': pressure, 'humidity': humidity, 'visibility': visibility, 'time_now': time_now})
        print(f'getting weather log for step number: {i+1} at {time_now}')
```
