```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://eg.freemeteo.com/weather/cairo/current-weather/location/?gid=360630&language=english&country=egypt'
response = requests.get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.find('div', attrs={'class': 'temp'})
```

```python
temp = soup.find('div', attrs={'class': 'temp'}).get_text()
temp
```

```python
temp.split('°')
```

```python
temp.split('°')[0]
```

```python
temp = float(temp.split('°')[0])
temp
```

```python
soup.find('span', attrs={'class': 'white'}).find('b')
```

```python
soup.find('span', attrs={'class': 'white'}).find('b').get_text()
```

```python
soup.find('span', attrs={'class': 'white'}).find('b').get_text().strip()
```

```python
wind = float(soup.find('span', attrs={'class': 'white'}).find('b').get_text().strip())
wind
```

```python
stats = soup.find('div', attrs={'class': 'stats'}).get_text()
stats
```

```python
stats.split(' ')
```

```python
for el in stats.split(' '):
    if el.endswith('%'):
        humidity = el
    elif el.endswith('m'):
        visibility = el
    elif el.endswith('mb'):
        presure = el
```

```python
humidity
```

```python
humidity.split('%')
```

```python
humidity = float(humidity.split('%')[0])
humidity
```

```python
visibility
```

```python
visibility.split('m')
```

```python
visibility = float(visibility.split('m')[0])
visibility
```

```python
presure
```

```python
presure.split('mb')
```

```python
presure = float(presure.split('mb')[0])
presure
```
